```python
#Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import randint
import math
```

Step 1

```python
#Step1 Initialisation
#variables declaration
w1=[0.5]
w2=[0.1]
lr=0.01
theta=[-0.1]
w=[w1,w2]
x1=[0,0,1,1]
x2=[0,1,0,1]
d_out=[0,1,1,1]
or_tab=[x1,x2]
p=0
epoch=150
e=[]
e_total=[]
y=[]
w1[0]
```

```python
for k in range(epoch): #Step4 Iteration from step2
    for i in range(len(x1)):
        #Step2 activation
        #calcul y
        y_temp=x1[i]*w1[p] + x2[i]*w2[p] + theta[p]
        y.append(y_temp)
        #Step3 weight training
        #update weight
        e.append(d_out[i]-y[p])
        w1.append(w1[p]+lr*x1[i]*e[p])
        w2.append(w2[p]+lr*x2[i]*e[p])
        theta.append(theta[p]+lr*theta[p]*e[p])
        p=p+1
    e_total.append((0.5)*(sum(e)*sum(e)))
    p=0
    e=[]
    y=[]
    w1_last=w1[-1]
    w1=[]
    w1.append(w1_last)
    w2_last=w2[-1]
    w2=[]
    w2.append(w2_last)
    theta_last=theta[-1]
    theta=[]
    theta.append(theta_last)
print(e_total)
```

```python
print("Final weights: "+str(w1)+" & "+str(w2))
plt.plot(range(150),e_total)
plt.show()
```
